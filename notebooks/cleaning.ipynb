```python
import numpy as np
import pandas as pd
import vaex
```

```python
# Importing the raw data into vaex dataframes

quotes_2015 = vaex.open('/content/drive/MyDrive/Q-Bank/DATA_HDF5/quotes-2015/*.hdf5')
quotes_2016 = vaex.open('/content/drive/MyDrive/Q-Bank/DATA_HDF5/quotes-2016/*.hdf5')
quotes_2017 = vaex.open('/content/drive/MyDrive/Q-Bank/DATA_HDF5/quotes-2017/*.hdf5')
quotes_2018 = vaex.open('/content/drive/MyDrive/Q-Bank/DATA_HDF5/quotes-2018/*.hdf5')
quotes_2019 = vaex.open('/content/drive/MyDrive/Q-Bank/DATA_HDF5/quotes-2019/*.hdf5')
quotes_2020 = vaex.open('/content/drive/MyDrive/Q-Bank/DATA_HDF5/quotes-2020/*.hdf5')
```

```python
quotes_2020.head()
```

```python
# dropping quotes with only 'None' speakers

df15 = quotes_2015[quotes_2015.speaker!='None']
df16 = quotes_2016[quotes_2016.speaker!='None']
df17 = quotes_2017[quotes_2017.speaker!='None']
df18 = quotes_2018[quotes_2018.speaker!='None']
df19 = quotes_2019[quotes_2019.speaker!='None']
df20 = quotes_2020[quotes_2020.speaker!='None']
```

```python
# function to keep only the most likely speaker

def keep_first_qid(qids):
    end_id = str(qids).find("\'",2)
    return str(qids)[2:end_id]
```

```python
# function to keep first probas

def keep_top_proba(probas):
  start_id = str(probas).find('0')
  end_id = str(probas).find("\'", start_id)
  if (start_id==-1 | end_id==-1):
    return probas
  return probas[start_id:end_id]
```

```python
# keep QID only for most likely speaker 

df15['qids'] = df15.qids.apply(keep_first_qid)
df16['qids'] = df16.qids.apply(keep_first_qid)
df17['qids'] = df17.qids.apply(keep_first_qid)
df18['qids'] = df18.qids.apply(keep_first_qid)
df19['qids'] = df19.qids.apply(keep_first_qid)
df20['qids'] = df20.qids.apply(keep_first_qid)
```

```python
# keep probas only for most likely speaker

df15['probas'] = df15.probas.apply(keep_top_proba)
df16['probas'] = df16.probas.apply(keep_top_proba)
df17['probas'] = df17.probas.apply(keep_top_proba)
df18['probas'] = df18.probas.apply(keep_top_proba)
df19['probas'] = df19.probas.apply(keep_top_proba)
df20['probas'] = df20.probas.apply(keep_top_proba)
```
